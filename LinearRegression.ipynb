```python
#Step 1: Import Libraries
#Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#Step 2: Import data

california = fetch_california_housing()
df = pd.DataFrame(data=california.data, columns=california.feature_names)
df['Precio'] = california.target

# Seleccionar la variable independiente

X = df[['MedInc']] # MedInc (Ingresos medios en la zona)
y = df['Precio']
df.head()
```

```python
# @title Median Income vs. House Price

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['MedInc'], df['Precio'], alpha=0.5)
plt.xlabel('Median Income')
plt.ylabel('House Price')
_ = plt.title('Median Income vs. House Price')
```

```python
X.head()
```

```python
#Step 3: Preprocessing
#Data visualization
df.plot(kind='scatter', x="MedInc", y="Precio")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Step 4: Model selection, setup and training
#Select linear regression model
model= sklearn.linear_model.LinearRegression()
```

```python
# Train model
#model.compile(metrics=['accuracy'])
model.fit(X, y)
```

```python
#Step 5: Test the model

# Making predictions with the trained model
# Normalize the income by dividing by 10,000 as in the original data set.
X_one = [[120000 / 10000]]  # Income 1
X_two = [[95000 / 10000]]   # Income 2
X_three = [[65000 / 10000]] # Income 3
X_four = [[45000 / 10000]]  # Income 4
X_five = [[30000 / 10000]]  # Income 5

pred_one = model.predict(X_one)
pred_two = model.predict(X_two)
pred_three = model.predict(X_three)
pred_four = model.predict(X_four)
pred_five = model.predict(X_five)

# Print predictions
print("Income forecast 1 ($120,000):", pred_one)
print("Income forecast 2 ($95,000):", pred_two)
print("Income forecast 3 ($65,000):", pred_three)
print("Income forecast 4 ($45,000):", pred_four)
print("Income forecast 5 ($30,000):", pred_five)
```
